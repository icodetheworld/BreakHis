```python

def get_pixel(img, center, x, y):     
    new_value = 0
    try: 
        if img[x][y] >= center: # center pixel is <= neighbour, then neighbour value = 1, otherwise 0
            new_value = 1
    except: 
        pass # pass through boundary condition
    return new_value

# Function for calculating LBP 
def lbp_calculated_pixel(img, x, y): 
    center = img[x][y] # Center pixel
    val_ar = [] 
    val_ar.append(get_pixel(img, center, x-1, y-1))     # top_left 
    val_ar.append(get_pixel(img, center, x-1, y))       # top 
    val_ar.append(get_pixel(img, center, x-1, y + 1))   # top_right 
    val_ar.append(get_pixel(img, center, x, y + 1))     # right 
    val_ar.append(get_pixel(img, center, x + 1, y + 1)) # bottom_right 
    val_ar.append(get_pixel(img, center, x + 1, y))     # bottom 
    val_ar.append(get_pixel(img, center, x + 1, y-1))   # bottom_left 
    val_ar.append(get_pixel(img, center, x, y-1))       # left
    power_val = [1, 2, 4, 8, 16, 32, 64, 128] # power of 2
    val = 0
    for i in range(len(val_ar)): 
        val += val_ar[i] * power_val[i] 
    return val # New val for center pixel is passed

def LBP_fn(path):
    img_bgr = cv2.imread(path, 1) 
    height, width, _ = img_bgr.shape 
    
    # We need to convert RGB image # into gray one because gray # image has one channel only. 
    img_gray = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2GRAY) 
    # Create a numpy array as # the same height and width # of RGB image 
    img_lbp = np.zeros((height, width), np.uint8) 
    for i in range(0, height): 
        for j in range(0, width): 
            img_lbp[i, j] = lbp_calculated_pixel(img_gray, i, j)
        m=100/height
        sys.stdout.write('\r')
        sys.stdout.write("[%s] %d%%" % ('='*round(i*m), m*i))
        sys.stdout.flush()
    plt.imshow(img_bgr) 
    plt.show() 
    plt.imshow(img_lbp, cmap ="gray") 
    plt.show() 
    print("LBP Program is finished") 
```

```python
def CLBP_fn(path):
    try:
        path = str(path)
        img_bgr = cv2.imread(path, 1) 
        # We need to convert RGB image # into gray one because gray # image has one channel only. 
        img_gray = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2GRAY) 
        numpoint = 24
        radius = 8
        eps=1e-7
        lbp = feature.local_binary_pattern(img_gray, numpoint,radius, method="uniform")
        (hist, _) = np.histogram(lbp.ravel(),bins=np.arange(0, numpoint + 3),range=(0, numpoint + 2))
        hist = hist.astype("float")
        hist /= (hist.sum() + eps)
        return list(hist)
    except Exception as e: 
        print(e)
```
